```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import json
```

```python
# Create an instance of MongoClient.
mongo = MongoClient(port=27017)
```

```python
# Create the database.
db_name = "uk_food"

dblist = mongo.list_database_names()

if db_name in dblist:
    print(f"{db_name} already exists.")
    print(f"Getting rid of the old and starting anew.")
    mongo.drop_database(db_name)

    db = mongo[db_name]
else:
    db = mongo[db_name]
```

```python
# Confirm the new database was created.
print(mongo.list_database_names())

# Note: this does not show the database until there is a collection with at least one document inside of it.
# But... it is working.
```

```python
# Create a collection.
col_name = "establishments"

col_list = db.list_collection_names()

if col_name in col_list:
    print(f"{col_name} already exists.")
    print(f"Getting rid of the old and starting anew.")
    db.drop_collection(col_name)

    coll = db[col_name]
else:
    coll = db[col_name]
```

```python
# Review the collections in our new database.
print(db.list_collection_names())

# Note: like above, this will not show the collection until there is at least a single document inside of it.
# But... still working.
```

```python
# Fill our collection using our json.
filepath = "../submission/Resources/establishments.json"

# Open the json.
with open(filepath, encoding = "utf8") as file:
    data = json.load(file)

    # Check if single or multiple entries.
    # Multiple.
    if isinstance(data, list):
        coll.insert_many(data)
    # Single.
    else:
        coll.insert_one(data)
```

```python
# Review a document in our new database.
pprint(coll.find_one())
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
q = {}
upd = [{
    "$set": {
        "geocode.latitude": {"$toDouble": "$geocode.latitude"},
        "geocode.longitude": {"$toDouble": "$geocode.longitude"}
    }
}]

coll.update_many(q, upd)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null.
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]

q = {
    "RatingValue": {"$in": non_ratings}
}

upd =  [ {'$set':{ "RatingValue" : None}} ]

coll.update_many(q, upd)
```

```python
# Change the data type from String to Integer for RatingValue.
q = {
    "RatingValue": {"$ne": None} 
}

upd = [{
    "$set": {
        "RatingValue": {"$toInt": "$RatingValue"}
    }
}]

coll.update_many(q, upd)
```

```python
# Check that the coordinates and rating value are now numbers.
results = coll.find({}).limit(5)

for result in results:
    print(f"Longitude: {result['geocode']['longitude']} -- Type: {type(result['geocode']['longitude'])}")
    print(f"Latitude: {result['geocode']['latitude']} -- Type: {type(result['geocode']['latitude'])}")
    print(f"Rating: {result['RatingValue']} -- Type: {type(result['RatingValue'])}")
```
